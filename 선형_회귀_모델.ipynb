# 선형 회귀 모델

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 공부 시간 x와 성적 y의 넘파이 배열을 만들기
x = np.array([2, 4, 6, 8])
y = np.array([81, 93, 91, 97])
```

```python
# 데이터의 분포를 그래프로
plt.scatter(x, y)
plt.show()
```

```python
# 기울기 a의 값과 절편 b의 값을 초기화
a = 0
b = 0
```

```python
# 학습률 정하기
lr = 0.03
```

```python
# 몇 번 반복될지 설정
epochs = 2001
```

```python
# x 값이 총 몇 개인지 세기
n = len(x)
```

```python
# 경사 하강법 시작
for i in range(epochs): # 에포크 수만큼 반복
    y_pred = a * x + b # 예측 값을 구하는 식
    error = y - y_pred # 실제 값과 비교한 오차를 error에 대입

    a_diff = (2/n) * sum(-x * (error)) # 오차 함수를 a로 편미분한 값
    b_diff = (2/n) * sum(-(error)) # 오차 함수를 b로 편미분한 값

    a = a - lr * a_diff # 학습률을 곱한 값만큼 기존의 a값을 빼줌
    b = b - lr * b_diff # 학습률을 곱한 값만큼 기존의 b값을 빼줌

    if i % 100 == 0: # 100번 반복될 때마다 현재의 a 값 b 값을 출력
      print('epoch=%.f, 기울기=%.04f, 절편=%.04f' % (i, a, b))
```

```python
# 앞서 구한 최종 a 값을 기울기 b 값을 y 절편에 대입해 그래프 그리기
y_pred = a * x + b
```

```python
# 그래프 출력
plt.scatter(x, y)
plt.plot(x, y_pred, color = 'red')
plt.show()
```

# 다중 선형 회귀 모델

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
# 공부 시간 x1과 과외 시간 x2, 성적 y의 넘파이 배열을 만든다.
x1 = np.array([2, 4, 6, 8])
x2 = np.array([0, 4, 2, 3])
y = np.array([81, 93, 91, 97])
```

```python
# 데이터의 분포를 그래프로 나타낸다
fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')
ax.scatter(x1, x2, y)
plt.show()
```

```python
# 기울기 a의 값과 절편 b의 값을 초기화한다.
a1 = 0
a2 = 0
b = 0
```

```python
# 학습률을 정한다.
lr = 0.01
```

```python
# 몇 번 반복?
epochs = 2001
```

```python
# x의 값이 총 몇 개인지 센다.
n = len(x1) # x1과 x2의 수가 같으므로 x1만 셈
```

```python
# 경사 하강법
```

```python
for i in range(epochs): # 에포크 수만큼 반복
    y_pred = a1 * x1 + a2 * x2 + b # 예측 값을 구하는 식
    error = y - y_pred # 실제 값과 비교한 오차를 error로 놓음

    a1_diff = (2/n) * sum(-x1 * (error)) # 오차 함수를 a1로 편미분한 값
    a2_diff = (2/n) * sum(-x2 * (error)) # 오차 함수를 a2로 편미분한 값
    b_diff = (2/n) * sum(-(error)) # 오차 함수를 b로 편미분한 값

    a1 = a1 - lr * a1_diff # 학습률을 곱한 값만큼 기존의 a1값을 빼줌
    a2 = a2 - lr * a2_diff # 학습률을 곱한 값만큼 기존의 a2값을 빼줌
    b = b - lr * b_diff # 학습률을 곱한 값만큼 기존의 b값을 빼줌

    if i % 100 == 0:
      print('epoch=%.f, 기울기1=%.04f, 기울기2=%.04f, 절편=%.04f' % (i, a1, a2, b))
```

```python
print("실제 점수: ", y)
```

```python
print("예측 점수: ", y_pred)
```

# 텐서플로에서 실행하는 선형 회귀

```python
import numpy as np
import matplotlib.pyplot as plt

# 텐서플로의 케라스 API에서 필요한 함수들을 불러옵니다.
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
x = np.array([2, 4, 6, 8])
y = np.array([81, 93, 91, 97])
```

```python
model = Sequential()
```

```python
# 출력 값, 입력 변수, 분석 방법에 맞게끔 모델을 설정
model.add(Dense(1, input_dim=1, activation='linear'))
```

```python
# 오차 수정을 위해 경사 하강법(sgd)을, 오차의 정도를 판단하기 위해
```

```python
# 평균 제곱 오차(mse)를 사용한다.
model.compile(optimizer='sgd', loss='mse')
```

```python
# 오차를 최소화하는 과정을 500번 반복한다.
model.fit(x, y, epochs=500)
```

```python
plt.scatter(x, y)
plt.plot(x, model.predict(x), color = 'red')
plt.show()
```

```python
# 임의의 시간을 집어넣어 점수를 예측하는 모델을 테스트해 보겠습니다.
hour = 7
prediction = model.predict(np.array([hour]))
print("%.f시간을 공부할 경우의 예상 점수는 %.02f점입니다." % (hour, prediction))
```

# 텐서플로에서 실행하는 다중 선형 회귀

```python
import numpy as np
import matplotlib.pyplot as plt

# 텐서플로의 케라스 API에서 필요한 함수들을 불러 옴
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
x = np.array([[2, 0], [4, 4], [6, 2], [8, 3]])
y = np.array([81, 93, 91, 97])
```

```python
model = Sequential()
```

```python
# 입력 변수가 두 개(학습 시간, 과외 시간)이므로 input_dim에 2를 입력
model.add(Dense(1, input_dim=2, activation='linear'))
model.compile(optimizer='sgd', loss='mse')
```

```python
model.fit(x, y, epochs=500)
```

```python
# 임의의 학습 시간과 과외 시간을 집어넣어 점수를 예측하는 모델을 테스트해 보겠습니다.
hour = 7
private_class = 4
prediction = model.predict(np.array([[hour, private_class]]))

print("%.f시간을 공부하고 %.f시간의 과외를 받을 경우, 예상 점수는 %.02f점입니다." % (hour, private_class, prediction))
```
